```python
# 필요한 열 선택
cols = ['stdYymm', 'nationNm', 'mprcExipitmNm', 'imxprtSeNm', 'imxprtWt', 'imxprtDollarAmount']

df = df[cols]
df.head()
```

```python
df.columns = ['기준년월', '국가명', '수산물명', '수출입구분', '수출입중량', '수출입달러']
df.head()
```

```python
# 고유값 확인
df['수산물명'].nunique()
```

```python
df['수산물명'] = df['수산물명'].str.replace(pat='\(.+', repl='', regex=True) \
                            .str.replace(pat=',.+', repl='', regex=True) \
                            .str.replace(pat=' 기타.*', repl='', regex=True)
```

```python
df['수산물명'].nunique()
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['수출입달러'] = df['수출입달러'].astype(float)
df['수출입중량'] = df['수출입중량'].astype(float)
df.dtypes
```

```python
df.head()
```

```python
df['수출입달러'].sum()
```

```python
df.groupby('수출입구분')['수출입달러'].sum()
```

```python
cond = df['수출입구분'].eq('수출')
df[cond].groupby('수산물명')['수출입달러'].sum().sort_values(ascending=False).head(10)
```

```python
cond = df['수출입구분'].eq('수입')
df[cond].groupby('수산물명')['수출입달러'].sum().sort_values(ascending=False).head(10)
```

```python
pd.pivot_table(df, index='수산물명', columns='수출입구분', values='수출입달러', aggfunc='sum', fill_value=0,
                margins=True, margins_name='합계').sort_values(by='수출', ascending=False).to_excel('Pivot_Table_Marine.xlsx')
```

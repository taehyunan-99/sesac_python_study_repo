# Pandas

```python
import pandas as pd
import numpy as np
```

```python
li1 = [1, 2, 3]

ar1 = np.array(li1)
ar1 
# array([1, 2, 3])

sr = pd.Series(li1)
sr
# 0    1
# 1    2
# 2    3
# dtype: int64
```

```python
li1 = [1, 2.0, 3]

ar1 = np.array(li1)
ar1 
# array([1., 2., 3.])

sr = pd.Series(li1)
sr
# 0    1.0
# 1    2.0
# 2    3.0
# dtype: float64
```

```python
li1 = [1, 2.0, '3']

ar1 = np.array(li1)
ar1 
# array(['1', '2.0', '3'], dtype='<U32')

sr = pd.Series(li1)
sr
# 0      1
# 1    2.0
# 2      3
# dtype: object
```

```python
type(sr) # pandas.core.series.Series
sr.ndim # 1
sr.shape # (3,)
sr.size # 3
sr.dtype # dtype('O')
```

### 시리즈 원소별 자료형 확인

```python
for i in sr:
    print(type(i))
# <class 'int'>
# <class 'float'>
# <class 'str'>
```

### 시리즈의 비교 연산

```python
sr >= 1
# a    True
# b    True
# c    True
# dtype: bool
```

```python
for i in sr:
    print(i >= 1 and i < 3)
# True
# True
# False

sr >= 1 and sr < 3
# ValueError: The truth value of a Series is ambiguous. Use a.empty, a.bool(), a.item(), a.any() or a.all().
```

### 비트 연산자

```python
(sr >= 1) & (sr < 3)
# a     True
# b     True
# c    False
# dtype: bool
```

```python
(sr >= 1) | (sr < 3)
# a    True
# b    True
# c    True
# dtype: bool
```

```python
(sr >= 1) ^ (sr < 3)
# a    False
# b    False
# c     True
# dtype: bool
```

```python
~((sr >= 1) & (sr < 3))
# a    False
# b    False
# c     True
# dtype: bool
```

## 데이터프레임

### 데이터프레임 생성

```python
li2 = [[1, 2, 3], [4, 5.0, '6']]

# 넘파이 배열
np.array(li2)
# array([['1', '2', '3'],
#        ['4', '5.0', '6']], dtype='<U32')

# 데이터프레임
df = pd.DataFrame(li2)
df
#   0	1	2
# 0	1	2.0	3
# 1	4	5.0	6
```

```python
df.ndim # 2
df.shape # (2, 3)
df.index # RangeIndex(start=0, stop=2, step=1)
df.columns # RangeIndex(start=0, stop=3, step=1)

df.dtypes
# 0      int64
# 1    float64
# 2     object
# dtype: object

df.values
# array([[1, 2.0, 3],
#        [4, 5.0, '6']], dtype=object)

df.info()
# <class 'pandas.core.frame.DataFrame'>
# RangeIndex: 2 entries, 0 to 1
# Data columns (total 3 columns):
#  #   Column  Non-Null Count  Dtype  
# ---  ------  --------------  -----  
#  0   0       2 non-null      int64  
#  1   1       2 non-null      float64
#  2   2       2 non-null      object 
# dtypes: float64(1), int64(1), object(1)
# memory usage: 180.0+ bytes
```

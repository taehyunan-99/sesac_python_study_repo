## for 반복문

```python
menu = ['짜장면', '탕수욕', '깐풍기', '짬뽕밥', '전가복', '샥스핀']

for item in menu:
    print(item, '시킬까요?')
# 짜장면 시킬까요?
# 탕수욕 시킬까요?
# 깐풍기 시킬까요?
# 짬뽕밥 시킬까요?
# 전가복 시킬까요?
# 샥스핀 시킬까요?
```

### 반복문에 조건문 추가

```python
for item in menu:
    print(item, '시킬까요?')
    if item in ['짜장면', '짬뽕밥']:
        print('-> 요리부터 주문합시다!')
    print('-> 다음 메뉴는 뭔가요?\n')
```

### 반복문 제어 - continue

```python
for item in menu:
    print(item, '시킬까요?')
    if item in ['짜장면', '짬뽕밥']:
        continue
    print('-> 다음 메뉴는 뭔가요?\n')
```

### 반복문 제어 - break

```python
for item in menu:
    print(item, '시킬까요?')
    if item in ['전가복', '샥스핀']:
        break
    print('-> 다음 메뉴는 뭔가요?\n')
```

### enumerate

```python
# enumerate() : 이터러블 객체의 인덱스와 원소를 튜플로 묶어서 반환
for idx, value in enumerate(menu):
    print(f'{idx + 1}번째 메뉴는 {value}입니다.')
# 1번째 메뉴는 짜장면입니다.
# 2번째 메뉴는 탕수욕입니다.
# 3번째 메뉴는 깐풍기입니다.
# 4번째 메뉴는 짬뽕밥입니다.
# 5번째 메뉴는 전가복입니다.
# 6번째 메뉴는 샥스핀입니다.
```

```python
# 여러개의 이터러블 객체들에서 같은 인덱스 원소를 튜플로 묶어서 차례대로 반환
price = [10000, 20000, 28000, 11000, 60000, 60000]

for i, j in zip(menu, price):
    print(f'{i}의 가격은 {j:,}원 입니다.')
# 짜장면의 가격은 10,000원 입니다.
# 탕수욕의 가격은 20,000원 입니다.
# 깐풍기의 가격은 28,000원 입니다.
# 짬뽕밥의 가격은 11,000원 입니다.
# 전가복의 가격은 60,000원 입니다.
# 샥스핀의 가격은 60,000원 입니다.
```

```python
for i, v in enumerate(menu, start=1):
    print(f'{i+1}번째 메뉴는 {v}입니다.')
# 2번째 메뉴는 짜장면입니다.
# 3번째 메뉴는 탕수욕입니다.
# 4번째 메뉴는 깐풍기입니다.
# 5번째 메뉴는 짬뽕밥입니다.
# 6번째 메뉴는 전가복입니다.
# 7번째 메뉴는 샥스핀입니다.
```

```python
for (idx, vlaue), p in zip(enumerate(menu), price):
    print(f'{idx + 1}번째 메뉴는 {vlaue}이고, 가격은 {p:,}원 입니다.')
# 1번째 메뉴는 짜장면이고, 가격은 10,000원 입니다.
# 2번째 메뉴는 탕수욕이고, 가격은 20,000원 입니다.
# 3번째 메뉴는 깐풍기이고, 가격은 28,000원 입니다.
# 4번째 메뉴는 짬뽕밥이고, 가격은 11,000원 입니다.
# 5번째 메뉴는 전가복이고, 가격은 60,000원 입니다.
# 6번째 메뉴는 샥스핀이고, 가격은 60,000원 입니다.
```

## 사용자 정의 함수

```python
def bmi(hgt, wgt):
    bmi = wgt / (hgt / 100) ** 2
    return bmi

# 함수 괄호 안에 매개변수와 전달인수를 등호로 연결하는 코드
bmi(hgt=178, wgt=70)
# 함수 괄호 안에 전달인수만 입력한 코드(순서가 바뀌면 결과도 바뀜)
# 매개변수를 추가할 수 없는 인수도 있고, 매개변수를 생략하면 안되는 인수도 있음
bmi(178, 70)
```

```python
# 기본값 및 독스트링 추가
# Hinting
# 전달인수의 기본값이 없는 매개변수는 왼쪽에, 있는 매개변수는 오른쪽에 놓아야 함
def bmi(hgt, wgt:float=70) -> float:
    '''
    This function returns bmi from weight(kg) and height(cm)
    '''
    return wgt / (hgt / 100) ** 2
```

### 가변 매개변수 : 위치 인수

```python
def printArgs(*args):
    for i in args:
        print(i)

printArgs(1, 2, 3)
# 1
# 2
# 3
```

```python
# product()
# 지정한 수를 모두 곱하여 반환

def product(*args):
    x = 1
    for i in args:
        x *= i
    return x
```

```python
product(3, 5, 2, 7, 10)
```

```python
# 지정한 정수를 3으로 나눈 나머지를 리스트로 반환하는 함수

def product_2(*args):
    return [i % 3 for i in args]
```

```python
product_2(5, 6, 10, 11)
```

```python
def printKwargs(**kwargs):
    for k, v in kwargs.items():
        print(f'{k}: {v}')

printKwargs(메뉴='순대국', 가격=9000, 평점=4.5, 위치='창동')
# 메뉴: 순대국
# 가격: 9000
# 평점: 4.5
# 위치: 창동
```

```python
# 키워드 가변인수를 받아 과목별 점수의 평균 계산
def my_avg(**kwargs):
    return sum(kwargs.values()) / len(kwargs)
```

## 참조 바인딩

- 참조 : 변수가 객체를 가리키는 것
- 참조 바인딩 : 두 변수가 같은 객체를 참조하는 것
- 얕은 복사
- 참조 반복
- 깊은 복사

```python
a = [[1, 2], [3, 4]]
```

```python
# 참조 바인딩은 원본을 유지할 수 없음
b = a

b[0] = 9

print(b) # [9, [3, 4]]
print(a) # [9, [3, 4]]
```

### 얕은 복사

```python
a = [[1, 2], [3, 4]]
```

```python
# 변수 a를 얕은 복사 : 컨테이너만 새로 만들고 내부 구조는 그대로 유지
b = a.copy()
```

```python
# b의 첫 번째 원소를 변경하면 b만 변경
# b는 참조를 바꾸고 a는 유지
```

### 참조 반복

- 컨테이너를 새로 만들지만, 내부 구조를 그대로 유지한 상태로 반복

```python
a = [[1, 2], [3, 4]]
```

```python
b = a * 2

print(a) # [[1, 2], [3, 4]]
print(b) # [[1, 2], [3, 4], [1, 2], [3, 4]]
```

```python
b[0][0] = 9

print(a) # [[9, 2], [3, 4]]
print(b) # [[9, 2], [3, 4], [9, 2], [3, 4]]
```

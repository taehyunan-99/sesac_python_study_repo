## 산술 연산자

````python
a = 5

print('+ :', a + 2)
print('- :', a - 2)
print('* :', a * 2)
print('** :', a ** 2)
print('/ :', a / 2)
print('% :', a % 2)
print('// :', a // 2)
````

## 할당 연산자

````python
a = 6
print(a)

a += 1
print(a)

a -= 2
print(a)
````

## 비교 연산자

````python
print('a > 4 :', a > 4)
print('a >= 4 :', a >= 4)
print('a < 4 :', a < 4)
print('a <= 4 :', a <= 4)
print('a == 4 :', a == 4)
print('a != 4 :', a != 4)
````

## 논리 연산자
- not의 우선순위가 and, or 보다 높기 때문에 괄호가 없으면 결과가 달라질 수 있음
    - ```not a > 4 and b > 4``` : a > 4 -> b > 4 -> not a > 4 -> 전체 실행 = False
    - ```not b > 4 and a > 4``` : b > 4 -> a > 4 -> not b > 4 -> 전체 실행 = True

````python
a, b = 7, 3

print(a > 4 and b > 4)
print(a > 4 or b > 4)
print(not(a > 4 and b > 4))
````

## 문자열 연산자

- 파이썬에서 값(객체)을 콤마로 나열하면 튜플로 패킹
- 오른쪽 개수만큼 왼쪽에 변수를 나열하면 순서대로 할당

````python
obj = '홍길동', 3.2
name, rate = '홍길동', 3.2

print(obj)
print(name, rate)
# ('홍길동', 3.2)
# 홍길동 3.2
````

````python
# 문자열 결합 (+)
name + ' 고객님의 이자율은 ' + str(rate) + '%압니다.'
# '홍길동 고객님의 이자율은 1.256%압니다.'
````

````python
# 문자열 반복 (*)
# 문자열 반복은 정수만 가능
print((name + '고객님! ') * 3)
# 홍길동고객님! 홍길동고객님! 홍길동고객님! 
````

````python
# 멤버 연산자 (in)
# 부분 문자열 포함 여부
'홍' in name
# True
````

## 문자열 메서드

````python
sen4 = '\n 파이썬 \t 문자열 \t 다루기 \n'

print(sen4)
#
#  파이썬 	 문자열 	 다루기 
#
````

````python
# strip()
sen4.strip()
# '파이썬 \t 문자열 \t 다루기'
````

````python
# replace()
sen4.replace('\t', '')
# '\n 파이썬  문자열  다루기 \n'
````

````python
# method chaining
sen4.replace('\t', '').strip()
# '파이썬  문자열  다루기'
````

````python
# split()
sen4.split('\t')
# ['\n 파이썬 ', ' 문자열 ', ' 다루기 \n']
````

````python
strs = sen4.split()
strs
# ['파이썬', '문자열', '다루기']
````

````python
# join()

' '.join(strs)
# '파이썬 문자열 다루기'
````

````python
# join()
addr = ['서울특별시', '도봉구', '창동', '1']
' '.join(addr)
# '서울특별시 도봉구 창동 1'

mobile = ['010', '1234', '5678']
'-'.join(mobile)
# '010-1234-5678'
````

## 문자열 포맷팅

- %f 는 변수가 가리키는 값을 실수로 가져오며, 소수점 여섯째자리까지 표기
- 원하는 소수점 자리수를 추가할 수 있음

````python
name = '홍길동'
rate = 1.256

print(name + '님의 이자율은 ' + str(rate) + '%입니다.')

# 포맷 코드를 활용한 문자열 포맷팅
print('%s님의 이자율은 %.2f%%입니다.' %(name, rate))

# format() 메서드 활용한 문자열 포맷팅
print('{0}님의 이자율은 {1:.2f}%입니다.'.format(name, rate))

# f-string 활용한 문자열 포맷팅
print(f'{name}님의 이자율은 {rate:.2f}%입니다.')

# 홍길동님의 이자율은 1.256%입니다.
# 홍길동님의 이자율은 1.26%입니다.
# 홍길동님의 이자율은 1.26%입니다.
# 홍길동님의 이자율은 1.26%입니다.
````
